### EDA

**Task 5: Data Analysis**

**Universal Analysis**

```python
#Identitying outliers
df =data[data['price'] <1500]
sns.boxplot(data=df,y='price')
```

```python
#Price Distribution
sns.histplot(data=df, x='price', bins=100)
plt.figsize=(10,5)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
df.columns
```

```python
#Availability_365 Distribution
sns.histplot(data=df, x='availability_365')
plt.figsize=(10,5)
plt.title('Availability_365 Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
df.dtypes
```

**Features Engineering**

```python
#Price per bed
df['price_per_bed'] = df['price'] / df['beds']
df.head()
```

```python
#Average price for beds grouped by neighbourhood group
df.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=True)
```

```python
#Average price per bed grouped by neighbourhood group
df.groupby('neighbourhood_group')['price_per_bed'].mean().sort_values(ascending=True)
```

**Bi Variate Analysis**

```python
df.columns
```

```python
#Price dependency on neighbourhood group
sns.barplot(data=df, x='neighbourhood_group', y='price_per_bed', hue='room_type')
```

```python
#relation between price and number of reviews
plt.figure(figsize=(10, 5))
plt.title('Price and Review Dependency')
plt.xlabel('Number of Reviews')
plt.ylabel('Price')
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.show()
```

```python
df.dtypes
```

```python
sns.pairplot(data=df, vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'] , hue='room_type')
plt.show()
```

```python
#Geopolitical distribution of price
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type')
plt.title('Geopolitical Distribution of Price')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
#heatmape of correlation for numerical columns
corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365']].corr()
corr
plt.figure(figsize=(10, 5))
plt.title('Correlation Heatmap')

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```
